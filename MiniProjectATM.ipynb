### ATM Class

```python
#ATM Class
#It contains all the methods required

class ATM:
    def __init__(self):
        self.pin = pin
        self.balance = 0
        self.savings = 1000
        self.name = name
        self.verification(pin)
        
#name = Username entered by user
#pin = pin entered by user in driver program
#balance = Current Balance in Main Account
#savings = Current Savings
#verification = Function for verifying the user to give access to ATM


    def verification(self,pin):
        pin_verif = input("\nPlease re-enter your pin : ")     #Taking pin Input again for verification
        if len(pin_verif) == 4:
            if pin_verif == pin:                               #Checking whether both the pins match or not
                time.sleep(2)
                print("\nUser authorized\n")
                print("*"*36)
                print(f'\nWelcome {self.name}\n')
                self.selectionMenu()
            else:
                print("""Pin didn't match
Please verify again""")
                self.verification(pin)
                print("*"*36)
        else:
            print("Please enter valid PIN")
            self.verification(pin)
    
    
#Selection Menu = Function which contains all the features this ATM has
#Main Menu


    def selectionMenu(self):
        print("""
Please Choose what you want to do 
""")
        print("-"*36)
        time.sleep(2)
        print("""
1   >>>   Check Balance in your Account
        
2   >>>   Deposit Funds in your Account

3   >>>   Withdraw Funds from your Account

4   >>>   Transfer funds to Savings Account

5   >>>   Exit
""")
        time.sleep(1)
        choice = int(input("\nYour Choice : "))
        if (choice == 1):
            print("\nChecking Balance in your account .... \n")
            time.sleep(2)
            self.CheckBalance()

        elif (choice == 2):
            print("\nSetting up Depositor .... \n")
            time.sleep(2)
            self.DepositFunds()

        elif (choice == 3):
            print("\nSetting up Withdraw Mechanism .... \n")
            time.sleep(2)
            self.WithdrawFunds()

        elif (choice == 4):
            print("\nFetching Saving Account Details .... \n")
            time.sleep(2)
            self.TransferFunds()

        elif (choice == 5):
            print(f"""
             Thank You 
      Have a Nice Day {self.name}
#######   Exiting System   #######""")
            time.sleep(2)
            exit()
        else:
            print("Please enter valid choice")
            self.selectionMenu()
            

#Check Balance = Function for checking Balance in Savings and Main Account


    def CheckBalance(self):
        time.sleep(2)
        print("""Check Balance for :
1   >>>   Main Account
2   >>>   Savings Account
3   >>>   Return to Main Menu\n""")
        choice = int(input("Please Enter your Choice : "))
        if choice == 1:
            time.sleep(1)
            print(f"\nYour Current balance in Main Account is : {self.balance}\n")       #Checking Balance in Main Account
            self.CheckBalance()
        elif choice == 2:
            time.sleep(1)
            print(f"\nYour Current Savings are : {self.savings}\n")                       #Checking Balance in Savings
            self.CheckBalance()
        elif choice == 3:
            print("\nRedirecting .... \n")
            time.sleep(1)
            self.selectionMenu()
        else:
            print("\nPlease Enter Valid Choice\n")
            time.sleep(1)
            self.CheckBalance()
        print("*"*36)
        

#Deposit Funds = Function for Deposting Funds in Main Account


    def DepositFunds(self):
        Deposit_Amount = int(input("Enter the amount you want to Deposit : "))                                                          
        print(f"\nDepositing {Deposit_Amount} in your Account\n")                        #Depositing some amount in main balance
        time.sleep(2)
        self.balance = self.balance + Deposit_Amount                                     #Incrementing Main Account Balance
        print(f"\nAmount of {Deposit_Amount} successfully Credited in your account\n")
        print("*"*36)
        print(f"\nYour Current Balance is : {self.balance}\n")
        print("*"*36)
        self.selectionMenu()


#Withdraw Funds = Function for Withdrawing funds from Main Account


    def WithdrawFunds(self):
        Withdraw_Amount = int(input("Enter the amount you want to Withdraw : "))
        if self.balance <= 0:
            print("\nInsufficient Balance\n")                                                #Checking if main balance is 0 or lower 
            print(f"Your Available Balance is : {self.balance}\n")
            self.selectionMenu()
        elif Withdraw_Amount > self.balance and self.balance != 0:                           #Checking if Withdraw amount is more than Account Balance
            print("\nInsufficient Balance\n")
            print("Please re-enter the Withdraw Amount\n\nRedirecting ..... \n")
            print("*"*36)
            time.sleep(2)
            self.WithdrawFunds()
        else:
            print(f"\nWithdrawing {Withdraw_Amount} from your Account ..... \n")
            time.sleep(2)
            self.balance = self.balance - Withdraw_Amount                                     #Updating Account balance after withdrawal
            print(f"\nAmount of {Withdraw_Amount} successfully Debited from your account\n")
            print("*"*36)
            print(f"\nYour Current Balance is : {self.balance}\n")
        print("*"*36)
        self.selectionMenu()
    

#Transfer Funds = Transferring Amount from Main Acccount Balance to Savings


    def TransferFunds(self):
        Transfer_Amount = int(input("Enter the amount you want to transfer : "))
        if Transfer_Amount > self.balance:                                                      #Checking if Transfer Amount is more than Account Balance
            time.sleep(1)
            print("\nInsufficient Balance\n")
            time.sleep(2)
            print(f"Available Balance in Main Account is : {self.balance}")
            print("\nPlease Enter the amount Again\nRedirecting .... \n")
            time.sleep(2)
            self.TransferFunds()
        else:
            self.balance = self.balance - Transfer_Amount
            self.savings = self.savings + Transfer_Amount                                         #Updating Account Balance and Savings balance after transfer
            time.sleep(2)
            print("\nFunds have been successfully transferred\n")
            time.sleep(1)
            print(f"\nCurrent Balance in Main Account is : {self.balance}")
            time.sleep(1)
            print(f"\nBalance in Savings Account is : {self.savings}\n")
            print("*"*36)
            print("\nRedirecting .... ")
        self.selectionMenu()
        
```
